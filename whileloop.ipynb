```python
num=int(input())
evensum=0
oddsum=0
primesum=0
while num>0:
    digit=num%10
    if digit%2==0:
        evensum+=digit
    else:
        oddsum+=digit
    if digit==2 or digit==3 or digit==5 or digit==7:
        primesum+=digit
    num//=10
print(f'sum of even numbers in a number : {evensum}')
print(f'sum of odd numbers in a number  : {oddsum}')
print(f'sum of prime numbers in a number : {primesum}')
```
