```python
"""Triangle Validity and Area
Check if a triangle is valid using sides, then calculate its area.
"""
a=float(input('enter a vlaues'))
b=float(input('enter b values'))
c= float(input('enter c value'))
if a+b>c and a+c>b and b+c>a:
    s=(a+b+c)/2
    area=(s*(s-a)*(s-b)*(s-c)) ** 0.5
    print('area',area)
else:
    print('invalid sides for triangle')
```

```python
"""7. Password Strength Checker
Input password and check:
Minimum 8 characters
Contains digit and special character
"""
password=input('eneter your password')
length=len(password)
digits=0
special=0
if length>=8:
    for i in password:
        if i>='0' and i<='9':
            digits+=1
        elif not ((i>='A' and i<='Z') or (i>='a' and i>='z')):
            special+=1
    if digits>=1 and special>=1:
        print('you have a strong password')
    else:
        print('weak password add digits or special characters')
else:
    print('keep the password with manimum 8 charcaters')
    
```

```python
"""Number Pattern Check
Input a number and check if it is:
Prime
Composite
Neither
"""
n=int(input('enter a number'))
if n<=1:
    print('niether')
is_prime=True
for i in range(2,n):
    if n%i==0:
        is_prime=False
        print('not a prime number')
        break
if is_prime==True:
    print('prime')
if n%2==0 or n%3==0:
    print('composite number')            
```
