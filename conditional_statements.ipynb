```python
#conditonal statements:--->are used to make decisions,They help a program choose different actions depending on whether a condition is True or False.
"""
if statement:========
Runs a block of code only if the condition is True.
if-else statement:===========
Chooses between two options: one if the condition is True, another if itâ€™s False.
if-elif-else statement:===============
Checks multiple conditions in order, and executes the first one that is True.
Nested if:=================
An if inside another if, used when decisions depend on multiple levels of checking.
"""
a=int(input('entera '))
b=int(input('enter b'))
if a>b:
    print(f'{a} is greater than {b}')
else:
    print(f'{b} is greater than {a}')
```
