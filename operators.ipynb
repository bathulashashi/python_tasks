```python
#operator
""" operators is used to perform some operations between two variables"""
#arthematic operator(+,-,*,**,//,%) is used to perform mathematical operation

# '+'used to merge the two list into a sinlge list
a=[1,2,4,'shashi']
b=[4,5,6]
print(a+b)

# '+'is also used to merge two tuples into a single tuples
a=(1,2,4,'shashi')
b=(1,2,4)
print(a+b)

# below method is used to merge two sets into a single set
a={1,2,4,'shashi'}
b={4,5,6}
print(*a,*b)

#below is used to merge two dictionaries into a single dictionary
c={1:'a',2:'b'}
d={3:'c'}
print({**c,**d})

#"+" to concatenate the strings 
a='shashi'
b='chandra'
print(a+b)

#'+ ' is used to perform addition b/w two numbers it maybe int or float
a=2
b=2.5
print(a+b)

"""
'*' to multiply b/w 2 variables
"""

# * is used to multiply b/w two numbers
a=2
b=20
print(a*b)
c=2.0
print(a*c)

#genereates repeations
d='shashi'
print(d*a)

#list repeatation
e=[1,2,3]
print(e*a)

#tuples repeataion
f=(1,2,33,5)
print(f*a)

"""note--------------------------------------------------------------------------------
#multiplication betwwen two string is not possible but tuples are multiplicable with a integer
#mutliplication b/w two list is not possible but tuples are multiplicable with a integer
#mutliplication b/w two tuples is not possible but tuples are multiplicable with a integer
#* is not posssible b/w two sets is not [possible
--------------------------------------------------------------------------------------
"""
#'-'

#'-' operation is possible b/w two numbers it may be float or integers
a=21
b=10
print(a-b)
c=20.0
print(a-c)

#'-' is possible b/w two bool values and also possible with bool( integer and flaot)
d=True
e=False
print(d-e)
print(e-d)

#'/'-----> divison 
a=10
b=2.0
print(a/b)
c=5
print(c/b)

#floor division to get integer number
print(c//b)

#'**' is used only for the numbers
a=2
b=3
print(a**b)
c=2.9
print(c**a)
```
